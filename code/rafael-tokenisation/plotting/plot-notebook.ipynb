```python
# Plot vocab vs embedding
import numpy as np
from transformers import GPTNeoConfig, GPTNeoForCausalLM, RobertaConfig, RobertaForMaskedLM
from scipy.interpolate import make_interp_spline

def calculate_parameters_gpt(vocab_size, embedding_size):
    # Assuming a simple model where the total number of parameters is the product of vocab size and embedding size
    config_gpt = dict(
        # EMBEDDING PARAMETERS
        vocab_size              = vocab_size,   # number of tokens in the vocabulary 
        hidden_size             = embedding_size,      # embedding size (vector length) of each token 
        max_position_embeddings = 512,      # maximum sequence length (context window)
    
        # BLOCKS (ATTN & FFN)
        num_layers          = 2,                    # number of transformer blocks
        attention_types     = [[["global", "local"], 1]], # (GPT-Neo-specific) global and local attention 
        num_heads           = 4,                    # attention heads
        window_size         = 256,                  # (GPT-Neo-specific) for local attention 
        intermediate_size   = 1024,                 # size of 'up-projection' layer in FFN
    
        pad_token_id = 0,           # need to specify this for tokenizer interop between models
    )
    
    config_gpt = GPTNeoConfig(**config_gpt)
    gpt = GPTNeoForCausalLM(config=config_gpt)
    return gpt.num_parameters()

def calculate_parameters_bert(vocab_size, embedding_size):
    # Assuming a simple model where the total number of parameters is the product of vocab size and embedding size
    config_rob = dict(
    
        # EMBEDDING PARAMETERS
        vocab_size              = vocab_size,   
        hidden_size             = embedding_size,      
        # we add 1 as RoBERTa uses a special position embedding for the padding token (zero vector)
        max_position_embeddings = 512 + 1,
    
        # BLOCKS (of course naming is different in roberta :) )
        num_hidden_layers = 2,
        num_attention_heads = 4,
        intermediate_size=1024,                     
    
        pad_token_id = 0,
    )
    
    config_rob = RobertaConfig(**config_rob)
    rob = RobertaForMaskedLM(config=config_rob)
    return rob.num_parameters()

# Target number of parameters
target_parameters = 9_000_000

# Vocabulary sizes from 1000 to 25000 with a step of 50
vocab_sizes = np.arange(1000, 25001, 500)

# List to store the corresponding embedding sizes
embedding_sizes_gpt = []
embedding_sizes_bert = []

# Iterate through each vocabulary size for GPT
for i, vocab_size in enumerate(vocab_sizes):
    print(f"Iteration {i} of {len(vocab_sizes)}")
    closest_parameters = float('inf')
    best_embedding_size = None
    for embedding_size in np.arange(4, 1000, 28):
        params = calculate_parameters_gpt(vocab_size, embedding_size)
        if abs(params - target_parameters) < abs(closest_parameters - target_parameters):
            closest_parameters = params
            best_embedding_size = embedding_size
    embedding_sizes_gpt.append(best_embedding_size)

# Iterate through each vocabulary size for BERT
for i, vocab_size in enumerate(vocab_sizes):
    print(f"Iteration {i} of {len(vocab_sizes)}")
    closest_parameters = float('inf')
    best_embedding_size = None
    for embedding_size in np.arange(4, 1000, 28):
        params = calculate_parameters_bert(vocab_size, embedding_size)
        if abs(params - target_parameters) < abs(closest_parameters - target_parameters):
            closest_parameters = params
            best_embedding_size = embedding_size
    embedding_sizes_bert.append(best_embedding_size)

# Fit a 2nd degree polynomial to the data for both GPT and BERT
coefficients_gpt = np.polyfit(vocab_sizes, embedding_sizes_gpt, 2)
poly_gpt = np.poly1d(coefficients_gpt)
coefficients_bert = np.polyfit(vocab_sizes, embedding_sizes_bert, 2)
poly_bert = np.poly1d(coefficients_bert)

# Generate smooth curve values
vocab_sizes_smooth = np.linspace(vocab_sizes.min(), vocab_sizes.max(), 500)
embedding_sizes_gpt_smooth = poly_gpt(vocab_sizes_smooth)
embedding_sizes_bert_smooth = poly_bert(vocab_sizes_smooth)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(vocab_sizes_smooth, embedding_sizes_gpt_smooth, label='GPT-Neo', color='blue')
plt.plot(vocab_sizes_smooth, embedding_sizes_bert_smooth, label='RoBERTa', color="red")
plt.xlabel('Vocabulary Size')
plt.ylabel('Embedding Size')
# plt.title('Embedding Size vs Vocabulary Size for 9M Parameters')
plt.legend()
# plt.grid(True)

# Save the plot as a high-resolution PDF file
plt.savefig("plots/vocab_size_vs_embedding_size.pdf", format='pdf', bbox_inches='tight')

# Display the plot
plt.show()
```
